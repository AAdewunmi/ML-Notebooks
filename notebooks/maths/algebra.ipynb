### Functions

Functions are used to map from one mathematical object to another. In ML, it's important to understand the concept of function as we use them a lot. In fact, for a lot of ML concepts we are essentially tying functions together mapping inputs to outputs. Let's start with some basic concept of functions, then gradually make our way into the more common functions applied in ML.

Example of a squaring function:

$$
f(x) = x^2
$$

```python
# import your main libraries

import matplotlib.pyplot as plt
import numpy as np
```

```python
# simple squaring function
def f(x):
    return x**2

# test the function
x = 2
print(f(x))
```

Let's visualize what that looks like:

```python
# visualize using matplotlib

x = np.linspace(-10, 10, 100)
y = f(x)
plt.plot(x, y)
```

A constant function:

$$
f(x) = c
$$

with $c$ being a constant.

```python
# a constant function
def f(x):
    return 2 # constant

# test the function
x = 2

# visualize using matplotlib
x = np.linspace(-10, 10, 100)
y = [f(x_i) for x_i in x]

plt.plot(x, y)
```

A linear function:

$$
f(x) = mx + c
$$

, with $m$ being the slope and $c$ being the y-intercept -- both constants.

```python
# a linear function
def f(x, m, c):
    return m*x + c

# test the function
x = 2
m = 3
c = 4

# visualize using matplotlib
x = np.linspace(-10, 10, 100)
y = [f(x_i, m, c) for x_i in x]

plt.plot(x, y)
```

Polynomial functions:

$$
f(x) = a_0 + a_1x + a_2x^2 + \cdots + a_nx^n
$$

with $a_0, a_1, a_2, ..., a_n$ being constants. And $n$ is the degree of the polynomial.

```python
# a polynomial function
def f(x, a, b, c):
    # a, b, c are coefficients
    return a*x**2 + b*x + c

# test the function
x = 2
a = 8
b = 1
c = 8

# visualize using matplotlib
x = np.linspace(-10, 10, 100)
y = [f(x_i, a, b, c) for x_i in x]

plt.plot(x, y)
```

Logarithmic functions:

$$
f(x) = c\log_a x
$$

with $c$ and $a$ being constants, $log_a$ is the logarithm function with base $a$.

```python
# a logarithmic function
def f(x, a):
    return a*np.log(x)  # natural log

# test the function
x = 2
a = 3

# visualize using matplotlib
x = np.linspace(0.1, 10, 100)
y = [f(x_i, a) for x_i in x]

plt.plot(x, y)
```

Exponential functions:

$$
f(x) = c\cdot a^x
$$

with $c$ and $a$ being constants.

```python
# an exponential function
def f(x, a, c):
    return c*np.exp(a*x)

# test the function
x = 2
a = 3
c = 4

# visualize using matplotlib
x = np.linspace(-10, 10, 100)
y = [f(x_i, a, c) for x_i in x]

plt.plot(x, y)
```

### Function Roots

Check if a specific function has root. Below we are checking linear function which has a unique root of $x = -c/m$.

```python
# function to find root of a linear function

def f(x, m, c):
    return m*x + c

def has_root(x, m, c):
    return x == -c/m

# generate range of values that includes root of linear function
x = np.linspace(-10, 10, 100)

# add root to the list
x = np.append(x, -c/m)

# order the list of values
x = np.sort(x)

y = [has_root(x_i, m, c) for x_i in x]

# y to visualize
y = [f(x_i, m, c) for x_i in x]

# visualize using matplotlib, with red dots at roots
plt.plot(-c/m, f(-c/m, m, c), 'r.')
plt.plot(x, y)

# label root value with text x = -c/m
plt.text(-c/m, f(-c/m, m, c), 'x = -c/m'+str(-c/m))
```

### Transformations of Functions

Transformations is an important concept where you take the output of one function and put it through another function.

Shifts happen when you move the function up or down, left or right. See example below of a vertical shift:

```python
# function f(x) = x**2
def f(x):
    return x**2

# test the function
x = np.linspace(-4, 4, 100)

# visualize using matplotlib
plt.plot(x, f(x))

# plot function shifted by 1 unit vertically; use red
plt.plot(x, f(x)+1, 'r')
```

Now we can do a horizontal shift:

```python
# plot function shifted by 2 units horizontally; use green
x = np.linspace(-10, 10, 50)
plt.plot(x, x**2)
plt.plot(x-2, x**2, 'g')

# have fixed x and y scale
plt.xlim(-20, 20)
```

Scaling transformation example:

```python
# vertical scaling
x = np.linspace(-6, 6, 100)
plt.plot(x, x**2)
plt.plot(x, (x**2)/2, 'g')

plt.xlim(-6, 6)
plt.ylim(-2, 4)

plt.plot(x, x*0, 'k', alpha=0.2)
plt.plot(x*0, x, 'k', alpha=0.2)
```

```python
# horizontal scaling
x = np.linspace(-6, 6, 100)
plt.plot(x, x**2)
plt.plot(x, -(x**2)/2, 'g')

plt.xlim(-6, 6)
plt.ylim(-4, 4)

plt.plot(x, x*0, 'k', alpha=0.2)
plt.plot(x*0, x, 'k', alpha=0.2)
```
